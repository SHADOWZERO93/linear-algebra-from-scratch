## 1. Vectors & Scalars

A **scalar** is a single number — it has magnitude only.
Examples: temperature (23°C), mass (5 kg), speed (60 km/h).

A **vector** has both magnitude and direction.
Picture an arrow in space — its length encodes how much, its orientation encodes which way.

In 2D, a vector is written as **v** = (v₁, v₂).
In 3D, as **v** = (v₁, v₂, v₃).

The **magnitude** (length) of a vector **v** = (a, b) is:

|**v**| = √(a² + b²)

This is just the Pythagorean theorem applied to the components.

## 2. Vector Addition & Scalar Multiplication

These are the two fundamental operations in linear algebra.
Everything else is built on top of them.

### Addition — tip to tail
Place the tail of the second vector at the tip of the first.
The result is the arrow from the original tail to the new tip.

Algebraically, just add component-wise:

**u** + **v** = (u₁ + v₁, u₂ + v₂)

### Scalar Multiplication
Multiplying a vector by a scalar stretches or shrinks it.

c · **v** = (c·v₁, c·v₂)

- c > 1  → stretches the vector
- 0 < c < 1 → shrinks the vector
- c < 0  → flips the direction
- c = 0  → zero vector

### Key properties
- Commutativity: **u** + **v** = **v** + **u**
- Distributivity: c(**u** + **v**) = c**u** + c**v**

```python
# Given u = (2, −3) and v = (−1, 5), compute u+v, u−v, and 3u−2v.
u = np.array([2,-3])
v = np.array([-1,5])
s = u + v
print(s)
```

## 3. Dot Product — Derived from Law of Cosines

The dot product combines two vectors into a single scalar.
There are two ways to compute it — component formula and geometric formula.
The derivation below proves they are the same thing.

### Starting point — Law of Cosines
For a triangle with sides a, b, c and angle θ opposite side c:

c² = a² + b² − 2ab·cos(θ)

### Derivation

Place vectors **u** and **v** at the origin.
The vector connecting their tips is **u** − **v**, forming a triangle.

**Step 1** — Write Law of Cosines for this triangle:

|**u** − **v**|² = |**u**|² + |**v**|² − 2|**u**||**v**|cos(θ)

**Step 2** — Expand |**u** − **v**|² component-wise:

|**u** − **v**|² = (u₁−v₁)² + (u₂−v₂)²
               = u₁² − 2u₁v₁ + v₁² + u₂² − 2u₂v₂ + v₂²
               = |**u**|² + |**v**|² − 2(u₁v₁ + u₂v₂)

**Step 3** — Set both expressions equal:

|**u**|² + |**v**|² − 2(u₁v₁ + u₂v₂) = |**u**|² + |**v**|² − 2|**u**||**v**|cos(θ)

**Step 4** — Cancel |**u**|² + |**v**|² and divide by −2:

u₁v₁ + u₂v₂ = |**u**||**v**|cos(θ)

### Result

**u · v = u₁v₁ + u₂v₂ = |u||v|cos(θ)**

The left side is the computational formula.
The right side is the geometric meaning.

```python
# Compute the dot product of u = (2, 3) and v = (4, −1) using all three methods — np.dot, @, and np.sum(u*v). Verify all three give the same answer.
u = np.array([2, 3])
v = np.array([4, -1])
print(np.dot(u,v))
print(u@v)
print(np.sum(u*v))
```

### What the dot product tells you

The dot product measures alignment between two vectors.

| Angle | cos(θ) | Dot product | Meaning |
|-------|--------|-------------|---------|
| 0°    | 1      | maximum positive | same direction |
| 90°   | 0      | 0 | perpendicular |
| 180°  | −1     | maximum negative | opposite directions |

- Positive dot product → acute angle
- Zero dot product → perpendicular
- Negative dot product → obtuse angle

To find the magnitude of a vector we should apply the Pythagoras formula

```python
# Given v = (5, 12), compute its magnitude without using np.linalg.norm — using only np.sqrt and np.sum.
v = np.array([5,12])
magn = np.sqrt(np.sum(v**2))
print(magn)
#we can also use np.linalg.norm
magni = np.linalg.norm(v)
print(magni)
```

## 4. Unit Vectors

A unit vector is a vector with magnitude exactly 1.
It represents pure direction — no magnitude information.

We get a unit vector by dividing any vector by its magnitude:

**v̂** = **v** / |**v**|

**Proof that this works:**

|**v̂**| = √((v₁/|v|)² + (v₂/|v|)²)
       = √((v₁² + v₂²) / |v|²)
       = √(v₁² + v₂²) / |v|
       = |v| / |v|
       = 1 ✓

The key step uses √(a/b) = √a/√b, which holds when a ≥ 0 and b > 0.

```python
# Compute the unit vector of u = (3, 4). Then verify its magnitude equals exactly 1.0 by printing it.
u = np.array([3,4])
ne = u / np.linalg.norm(u)
print(ne)
l = np.array(ne)
magnitude = np.linalg.norm(l)
print(magnitude)
```

```python
# Write a function normalize(v) that takes any vector and returns its unit vector. Test it on (1, 1).
def unitvector(v):
    u = v / np.linalg.norm(v)
    return u

# Test on (1, 1)
v1 = np.array([1, 1])
print("Vector      :", v1)
print("Unit vector :", unitvector(v1))

# Test on (0, 3)
v2 = np.array([0, 3])
print("Vector      :", v2)
print("Unit vector :", unitvector(v2))

# Test on (-5, 12)
v3 = np.array([-5, 12])
print("Vector      :", v3)
print("Unit vector :", unitvector(v3))
```

## 5. Angle Between Vectors

Rearranging the dot product formula:

cos(θ) = (**u** · **v**) / (|**u**| |**v**|)

θ = arccos((**u** · **v**) / (|**u**| |**v**|))

**Important:** Always use np.clip before np.arccos.
Floating point errors can push cos(θ) just past ±1,
causing arccos to return nan.
np.clip(cos_theta, -1, 1) prevents this.

```python
# Find the angle between u = (1, 0) and v = (0, 1) in degrees. Does the answer match your earlier hand calculation?
u = np.array([3, 4])
v = np.array([-4, 3])
angle = (u@v) / (np.linalg.norm(u) * np.linalg.norm(v))
nangle = np.arccos(angle)
print(np.degrees(nangle))
```

## 6. Projections

The projection of **u** onto **v** is the shadow **u** casts
onto the line of **v** if light shone perpendicular to **v**.

### Scalar projection
The signed length of the shadow:

scalar projection = (**u** · **v**) / |**v**|

### Vector projection
The actual vector along **v** with that length:

vector projection = ((**u** · **v̂**)) · **v̂**

where **v̂** = **v** / |**v**| is the unit vector of **v**.

**Important:** Always multiply by **v̂**, never by **v**.
Multiplying by **v** instead of **v̂** stretches the result
by |**v**| and gives the wrong answer.

### Verification
The remainder vector **u** − (vector projection) must be
perpendicular to **v**. Check: their dot product should be 0.

```python
# Compute the scalar projection of u = (2, 1) onto v = (1, 3). 
u = np.array([2, 1])
v = np.array([1, 3])
scalarprojection = (u@v) / (np.linalg.norm(v))
print(scalarprojection)
```

```python
# Compute the vector projection of u = (3, 2) onto v = (3, 2). 
u = np.array([3, 2])
v = np.array([3, 2])
vhat = v / (np.linalg.norm(v))
vectorprojection = (((u@v) / np.linalg.norm(v))*vhat)
rem = u-vectorprojection
print(vectorprojection)
print(rem)
```

```python
# Write a function that takes two vectors and returns: magnitude of each, angle between them, dot product, and whether they are perpendicular (True/False). Test it on at least three pairs.
def vector_info(u, v):
    firstmagnitude = np.linalg.norm(u)
    secondmagnitude = np.linalg.norm(v)
    dotproduct = u @ v
    ang = dotproduct / (firstmagnitude * secondmagnitude)
    anglebetween = np.degrees(np.arccos(np.clip(ang, -1, 1)))
    is_perpendicular = np.isclose(dotproduct, 0)
    scalarprojection = dotproduct / np.linalg.norm(v)
    vhat = v / np.linalg.norm(v)
    vectorprojection = (u @ vhat) * vhat

    print("Magnitude of u     :", firstmagnitude)
    print("Magnitude of v     :", secondmagnitude)
    print("Dot product        :", dotproduct)
    print("Angle between      :", anglebetween, "degrees")
    print("Perpendicular      :", is_perpendicular)
    print("Scalar projection  :", scalarprojection)
    print("Vector projection  :", vectorprojection)
    print()

# Test pair 1 — general vectors
u1 = np.array([2, 3])
v1 = np.array([4, -1])
print("Pair 1 — u=(2,3), v=(4,-1)")
vector_info(u1, v1)

# Test pair 2 — perpendicular vectors
u2 = np.array([3, 4])
v2 = np.array([-4, 3])
print("Pair 2 — u=(3,4), v=(-4,3) ")
vector_info(u2, v2)

# Test pair 3 — parallel vectors
u3 = np.array([1, 2])
v3 = np.array([2, 4])
print("Pair 3 — u=(1,2), v=(2,4) ")
vector_info(u3, v3)
```

## 7. Key Takeaways

- A vector has both magnitude and direction — a scalar has only magnitude
- Unit vector = v / |v| — always has magnitude exactly 1
- Vector addition is tip-to-tail geometrically, component-wise algebraically
- Dot product = 0 means the vectors are perpendicular
- Dot product = |u||v| means the vectors are parallel
- Always use np.clip(-1, 1) before np.arccos to avoid floating point errors
- Always use np.isclose() instead of == when comparing floats
- Always project using v̂ (unit vector), never the original v
